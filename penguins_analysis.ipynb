# Penguins Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
peng = sns.load_dataset("penguins")
#peng columns= species', 'island', 'bill_length_mm', 'bill_depth_mm','flipper_length_mm', 'body_mass_g', 'sex'

#Relationship between bill depth and bill length of each species
bx=sns.jointplot(data=peng, x="bill_depth_mm", y="bill_length_mm", hue="species", kind="kde")
bx.fig.suptitle("Relationship between bill depth and bill length of each species", y=1.02) # Adjust y to avoid overlapping titles
plt.xlabel("Bill Bepth(mm)")
plt.ylabel("Bill Length(mm)")
```

```python
#Species of each island
p=sns.FacetGrid(peng, col="island", row="species")
p.map(sns.histplot, "body_mass_g", bins=30, kde=True)
plt.tight_layout()
plt.subplots_adjust(hspace=1, wspace=1)
```

```python
#Correlation between numerical variables
d=sns.heatmap(peng.select_dtypes(include=["number"]).corr(), annot=True)
d.set_title("Correlation between numerical variables")
```

```python
#Body mass by species

ax=sns.barplot(data=peng, x="species", y="body_mass_g", hue="species")
ax.set_title("Body mass by species")
plt.xlabel("Species")
plt.ylabel("Body Mass(mm)")
```

Conclusions:

*  Species with greater bill length have smaller bill depth, leading to the conclusion that the larger one is, the smaller the other will be.
*   The Adelie species is present on all three islands, while the Chinstrap species is present only on Dream Island and the Gentoo species is present only on Biscoe Island.

*   There is a positive correlation between body mass and flipper length, so penguins with greater body mass will have greater flipper length.
*   Gentoo penguins generally weigh more than Adelie and Chinstrap penguins; these two species generally weigh similarly.
